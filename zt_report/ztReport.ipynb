```python
import yaml

def prep_yaml(yaml, page_count = 3, max_line_length = 120, lines_per_page = 90, word_truncate_length = 10):
    pages = []
    
    lines = yaml.split('\n')
    total_lines = len(lines)
    page = ''
    page_line_count = 0
    line = None
    line_idx = 0
    indent_s = ''
    current_max_line_length = max_line_length
    first_split_line = True
    
    while line_idx < total_lines:
        line_to_add = ''
        if line is None:
            line = lines[line_idx]
            line_idx += 1
            first_split_line = True
            indent_s = ''
            current_max_line_length = max_line_length
            
        if len(line) < current_max_line_length:
            line_to_add = line
            line = None
        else: # split line
            line_to_add = line[0:current_max_line_length]
            line = line[current_max_line_length:]
            
            if first_split_line:
                indent = len(line) - len(line.lstrip())
                current_max_line_length = max_line_length - indent
                indent_s = ' '*indent
                first_split_line = False
            
            last_space_idx = line_to_add.rfind(' ')
            if last_space_idx < (current_max_line_length - word_truncate_length):
                line_to_add += '-'
            else:
                line = line_to_add[last_space_idx:] + ' ' + line
                line_to_add = line_to_add[0:last_space_idx]
                
            line_to_add = indent_s + line_to_add
        
        if page_line_count >= lines_per_page:
            pages.append(page)
            page_line_count = 0
            page = ''
            
            if len(pages) >= page_count:
                break
            
        page += line_to_add + '\n'
        page_line_count += 1
    
    if page_line_count > 0:
        pages.append(page)
        
    while len(pages) < page_count:
        pages.append('')
    
    return pages

phtml = ''
for i in range(0, len(props)):
    prop = props.iloc[i]
    if prop['id'] != leading_prop_id:
        continue

    pages = []
    if 'detail' in prop:
        yaml = yaml.dump(prop['detail'], sort_keys = False, allow_unicode=True)
        yaml_pages = prep_yaml(yaml, max_line_length = 160)
        prop_detail = {key.title():value for key, value in prop['detail'].items()}
        hierarchy = prop_detail['Hierarchy'].split('/')
        leaf = hierarchy[len(hierarchy) - 1]
        hierarchy.pop()
        if 'Summary_Country' in prop_detail:
            hierarchy[0] = prop_detail['Summary_Country']
        else:
            hierarchy[0] = 'USA'
        hierarchy = '%2F'.join(hierarchy)
        phtml += '<a href="{:}/path/{:}/issue/{:}" target="_parent">View Full Proposal Text</a>'.format(holon, hierarchy, leaf)

        for idx, yaml_page in enumerate(yaml_pages):
            if 'author' in prop['detail']:
                phtml += '<div style="color: #161515; font-size: 14px; line-height: 15px; letter-spacing: 0.03em; font-family: Poppins;"><h3>Proposal ID {:} by {:} on {:} page {:}</h3>'.format(prop['id'], prop['detail']['author']['name'], prop['date'], idx+1)
            else:
                phtml += '<div style="color: #161515; font-size: 14px; line-height: 15px; letter-spacing: 0.03em; font-family: Poppins;"><h3>Proposal ID {:} page {:}</h3>'.format(prop['id'], idx+1)
            phtml += '<pre style="font-size: 14px; line-height: 20px; font-family: Poppins;">{:}</pre>'.format(yaml_page)
            phtml += '<div style="page-break-after: always;"></div>'

        break
    else:
        phtml += '<h3>Proposal ID {:} page {:}</h3>'.format(prop['id'], idx+1)
        phtml += '<p style="font-size: 8pt; font-family: Open Sans;">Invalid yaml</p>'
        phtml += '<div style="page-break-after: always;"></div>'
display(HTML(phtml))
```
